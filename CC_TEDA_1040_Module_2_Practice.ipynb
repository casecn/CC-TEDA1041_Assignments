#### Question 1
Using matplotlib/seaborn, create a countplot that shows a count of how many of the companies in the data set are from each state.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('companies.csv')
df.head()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="state", data=df)
plt.title("Count of Companies by State")
plt.show()
```

```python
counts = df['state'].value_counts()
sorted_by_index = counts.sort_index()
print("\nvalue_counts sorted by index:")
print(sorted_by_index)
```

```python
df['state'] = df['state'].replace({'Ca': 'CA'})
```

```python
plt.figure(figsize=(10, 5))
count_plot = sns.countplot(x="state", data=df)
count_plot.set_title("Count of Companies by State", fontsize=14)
count_plot.set_xlabel('State', fontsize=12)
count_plot.set_ylabel('Count', fontsize=12)
plt.show()
```

#### Question 3
Using matplotlib/seaborn, create a bar chart that shows a sum of how many total job openings there are in each industry.

```python
counts = df['Industry'].value_counts()
sorted_by_index = counts.sort_index()
print("\nvalue_counts sorted by index:")
print(sorted_by_index)
```

```python
sum_of_openings = df[['Industry', 'job_openings']].groupby('Industry', as_index=False).sum()
sum_of_openings.head()
```

```python
plt.figure(figsize=(15, 5))
count_plot = sns.barplot(x="Industry", y="job_openings", data=sum_of_openings, width=0.6)
count_plot.tick_params(axis='x', rotation=45)
count_plot.set_title("Count of Jobs by Industry", fontsize=14)
count_plot.set_xlabel('Industry', fontsize=12)
count_plot.set_ylabel('Count', fontsize=12)
plt.show()
```
